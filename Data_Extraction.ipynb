```python
import time
import pandas as pd

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from bs4 import BeautifulSoup
```

```python
def to_get_Product_hyper_link(url):
    driver = webdriver.Chrome()  # Optional argument, if not specified will search path.

    driver.get(url)

    time.sleep(5) # Let the user actually see something!

    search_box = driver.find_element(By.ID, 'twotabsearchtextbox') # To Find the Search box

    search_box.send_keys('t shirt for men') # Enter the Search values

    search_box.submit()

    url_list=[]

    elements = driver.find_elements(By.XPATH, '//div[@data-component-type="s-search-result"]//span//span/a[@class="a-link-normal s-no-outline"][@href]')
    
    url_list.extend([element.get_attribute('href') for element in elements])

    driver.quit()

    return url_list
```

```python
url = 'https://www.amazon.in/'
hyper_links = to_get_Product_hyper_link(url)
hyper_links
```

```python
def to_get_Product_details(url):

    driver = webdriver.Chrome()
    driver.get(url)
    time.sleep(5)
    try:
        element = driver.find_element(By.XPATH, '//div[@id="centerCol"]')
        elements = driver.find_element(By.XPATH, '//div[@data-feature-name="productFactsDesktop"]/div')


        data = dict(productname = element.find_element(By.XPATH, './/h1').text,
                    productprice = element.find_element(By.XPATH, '//div[@id="centerCol"]/div/div/div/div/div/span/span[2]/span[2]').text,
                    discount = element.find_element(By.XPATH, '//div[@id="centerCol"]/div/div/div/div/div/span[2]').text,
                    Material_composition = element.find_element(By.XPATH, '//div[@data-feature-name="productFactsDesktop"]/div/div[1]/div[1]/div/div[2]/span/span').text,
                    Pattern = elements.find_element(By.XPATH, '//div[@data-feature-name="productFactsDesktop"]/div/div[1]/div[2]/div/div[2]/span/span').text,
                    Fit_type= elements.find_element(By.XPATH, '//div[@data-feature-name="productFactsDesktop"]/div/div[1]/div[3]/div/div[2]/span/span').text,
                    Sleeve_type= elements.find_element(By.XPATH, '//div[@data-feature-name="productFactsDesktop"]/div/div[1]/div[4]/div/div[2]/span/span').text,
                    Collar_style= elements.find_element(By.XPATH, '//div[@data-feature-name="productFactsDesktop"]/div/div[1]/div[5]/div/div[2]/span/span').text,
                    Length= elements.find_element(By.XPATH, '//div[@data-feature-name="productFactsDesktop"]/div/div[1]/div[6]/div/div[2]/span/span').text,
                    Country_of_Origin= elements.find_element(By.XPATH, '//div[@data-feature-name="productFactsDesktop"]/div/div[1]/div[7]/div/div[2]/span/span').text,
                    ProductLink = url)

        driver.quit()
    except NoSuchElementException as e:
        pass

    finally:
        driver.quit()

    return data
```

```python
details=[]

for i in range(len(hyper_links)):
    details.append(to_get_Product_details(hyper_links[i]))

df= pd.DataFrame(details)
df
```
